# Matplotlib (simple visualization library)

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(10,20)
```

# scatterplot

```python
plt.scatter(x,y,color='blue')
plt.xlabel('x-value')
plt.ylabel('y-value')
plt.title('simple scatterplot')
```

# plt plot 

```python

plt.plot(x,y)
```

```python
plt.plot(x,y,'r*--')
```

# subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r-')
plt.subplot(2,2,2)
plt.plot(x,y,'b*')
plt.subplot(2,2,3)
plt.plot(x,y,'g^')
```

# sin and cosine

```python
np.pi
```

```python
#sin
a=np.arange(0,4*np.pi,0.1)
b=np.sin(a)
plt.plot(a,b)
plt.show()
```

```python
#cosine

a=np.arange(0,4*np.pi,0.1)
b_cos=np.cos(a)
plt.plot(a,b)
```

# Barplot

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
```

# Histogram

```python
x=[1,82,93,74,55,96,46,73,89]
```

```python
plt.hist(x)
plt.title('Histogram')
```

# Box plot

```python
data=[np.random.normal(1,std,100)for std in range(1,4)]
data
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
```

# pie chart

```python
#data to plot

labels=['python','c++','java','ruby','scala']
sizes=[240,123,450,100,300]
colors=['gold','yellow','blue','green','pink']
explode=(0.1,0,0,0,0)

#plot

plt.pie(sizes,labels=labels,explode=explode,colors=colors,shadow=True)
plt.show()
```
